```python
%cd /data/lyna/histogram-resampling/
import pandas as pd
data=pd.read_csv('junk.csv')
data
```

```python
fig,ax=plt.subplots()
ax.plot(range(len(data.data)),data.data, marker="o")
plt.show()
plt.savefig('viz0.png')
```

```python
hist_sample=[]
n_replicas=10
#resampling with replacement, == Bootstrapping using pandas.Series.sample
# below is a random 100% (frac=1) sample of the data with replacement
# replacement means allow sampling of the same row more than once
histo_sample = pd.Series([data.sample(frac=1, replace=True) 
                         for i in range(n_replicas)])
histo_sample
```

```python
#histogram, determining bins
n_bins=13

bins=np.histogram(np.hstack(histo_sample[:]), bins=n_bins)[1]
print(bins)
```

```python
#collect resampled data
histos=np.zeros([n_replicas,n_bins])
for i in range(n_replicas): 
    histos[i,:]=np.histogram(histo_sample[i].data, bins)[0]
    print(histos[i,:])
    plt.bar(np.linspace( 0,n_bins-1,n_bins ),histos[i,:])
```

```python
#stats
lo,med,up = np.zeros(n_bins),np.zeros(n_bins),np.zeros(n_bins)
for i in range(n_bins):
    med[i]=np.mean( histos[:,i] )
    lo[i]=med[i] - np.percentile( histos[:,i],5 )
    up[i]=np.percentile( histos[:,i], 95 ) - med[i]
```

```python
#viz.
x=np.linspace( 0,n_bins-1,n_bins )
plt.errorbar( x, med, yerr=[lo,up],fmt='ko',ecolor=None )
plt.bar( x,med,width=0.5 )
plt.show()
plt.savefig('viz1.png')
```
